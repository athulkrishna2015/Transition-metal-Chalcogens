```python
rename_dict = {'Formula': 'formula', 'Band gap': 'target'}
df0 = df0.rename(columns=rename_dict)
# df.head()
```

```python
df1['target'] = 0
# df.head()
df1.columns = ['formula','target']
# df1.head()
```

```python
 df_f = df1.head(0)
```

```python
for i in range(len(df1.index)):
  i+=1
  df3 = df1.iloc[[i]] # extract i th row df
  df = pd.concat([df3,df0], ignore_index=True)
  X_unscaled, y, formulae, skipped = generate_features(df, elem_prop='oliynyk', drop_duplicates=False, extend_features=False, sum_feat=True)
  scaler = StandardScaler()
  X = scaler.fit_transform(X_unscaled)
  y_predict = model.predict(X)
  # y_predict
  df['target'] = y_predict
  # df.sort_values(by=['target'], ascending=False)
  df_f = pd.concat([df_f,df.head(1)], ignore_index=True)
```
